```python
from collections import Counter
cuisine_list = [cuisine.strip() for cuisine in df["Cuisines"] for cuisine in cuisine.split(",")]
cuisine_list
```

```python
cuisine_counts = Counter(cuisine_list)
cuisine_counts
top_most_cuisine = cuisine_counts.most_common(3)
top_most_cuisine
```

```python
total_restaurants = len(df)
```

```python
cuisine_percentage = [round((count/total_restaurants)*100,2) for cuisine, count in top_most_cuisine]
cuisine_percentage
```

```python
top_cuisine_percentage = [(cuisine,count,round((count/total_restaurants)*100,2) )for cuisine, count in top_most_cuisine]
top_cuisine_percentage
```
